```python
# libraries
import codecs
import os
import json
import tqdm
from collections import Counter
from nltk import FreqDist
import pandas as pd
import numpy as np
import itertools
from hazm import *
```

```python
# read data
# option shows that all the data are stored in a json or each movie stored in seperate file.
DIRNAME = 'zoomg-movies-review'
option = 2

if option == 1:
    movies_dict = {}
    for file_name in os.listdir(DIRNAME):
        with open(f"{DIRNAME}/{file_name}", "r", encoding="utf-8") as text_file:
            movies_dict[file_name.replace('.txt', '')] = json.loads(text_file.read())
elif option == 2:
    with open(f"{DIRNAME}/movies.json", "r", encoding="utf-8") as text_file:
        movies_dict = json.loads(text_file.read())
```

### normalization

```python
# normalization
from hazm import *

normalizer = Normalizer()

normalized_movies_dict = {}
for movie, value in tqdm.tqdm(movies_dict.items()):
    review_text = value['text']
    normalized_movies_dict[movie] = {"writer" : normalizer.normalize(value['writer']), "text":normalizer.normalize(review_text)}
```

<div style="direction:rtl;font-family:XB Nazanin;font-size:16px;">
تابعی برای درست کردن دیتا‌فریم مربوط به توکنایز‌کردن و تحلیل بسامد: 
</div>

```python
def create_dataframe_freq(movies_dict, stopwords=[]):
    movies_tokens = {}
    for movie, value in tqdm.tqdm(movies_dict.items()):
        movies_tokens[(movie,value['writer'])] = FreqDist([_ for _ in word_tokenize(value['text']) if _ not in stopwords]).most_common(50)
    return pd.DataFrame(movies_tokens)
```

<div style="direction:rtl;font-family:XB Nazanin;font-size:16px;">
اطلاعاتی از آن دریافت می‌کنیم این است که معمولا در نقد و بررسی فیلم‌ها خود کلمه فیلم بیشترین تعداد به‌کار رفته است.
    یا مثلا در سریال‌ها معمولا کلمه فصل بیشترین تکرار را داشته‌است.
 اطلاعات دیگری هم می شود از آن دریافت کرد.  
<br>
    برای مثال سریال ted lasso راجع به مربی تیم فوتبالی است و همانطور که می‌بینید در نقد آن کلمه "تیم" بسیار به کار رفته است.
    از فیلم‌های دیگر هم همچین نکاتی می‌توان استخراج کرد.
</div>

<div style="direction:rtl;font-family:XB Nazanin;font-size:16px;">
اما برای تحلیل بهتر سعی می‌کنیم همین تحلیل‌ها را به جای هر فیلم بر اساس نویسنده‌ها انجام دهیم تا اطلاعات بهتری و بیشتری در اختیار داشته باشیم و آن‌ها را تا حدودی مقایسه کنیم.
    <br>
دوباره تحلیل بسامد را این‌بار به تفکیک نویسنده‌ها انجام می‌دهیم.
</div>

```python
writers = set(normalizer.normalize(_['writer']) for _ in normalized_movies_dict.values())
writers_texts = {writer : [] for writer in writers}

for item in tqdm.tqdm(normalized_movies_dict.values()):
    writers_texts[normalizer.normalize(item['writer'])].extend(_ for _ in word_tokenize(item['text']) if _ not in total_stop_words)

tokens_categorized_by_writers = {writer: FreqDist(tokens).most_common(50) for writer,tokens in tqdm.tqdm(writers_texts.items())} 
freq_analysis_by_writers = pd.DataFrame(tokens_categorized_by_writers)
```

<div style="direction:rtl;font-family:XB Nazanin;font-size:16px;">
حال سعی می‌کنیم به کمک 
    Lemmatization/Stemming
    اطلاعات ما بهتر و کاربردی‌تر شوند و مثلا کلمات "فیلم" و "فیلمی" یکی شوند.
</div>

```python
stemmer = Stemmer()
lemmatizer = Lemmatizer()

def get_lemma_set(tok, opt=1):
    if opt ==1:
        return stemmer.stem(tok)
    if opt ==2:
        return lemmatizer.lemmatize(tok)
    
```

```python
opt = 2
writers_text_after_lemmatize = {}

for writer, tokens in writers_texts.items():
    writers_text_after_lemmatize[writer] = [get_lemma_set(token, opt) for token in tokens]

tokens_categorized_by_writers = {writer: FreqDist(tokens).most_common(50) for writer, tokens in tqdm.tqdm(writers_text_after_lemmatize.items())} 
freq_analysis_by_writers = pd.DataFrame(tokens_categorized_by_writers)    
```

```python
writers_num_movies_reviewed_dict = {}
for (name, number) in writers_num_movies_reviewed.most_common():
    writers_num_movies_reviewed_dict[name] = number
```

```python
writers_sentences_data = {writer : [] for writer in writers}
for item in tqdm.tqdm(normalized_movies_dict.values()):
    writers_sentences_data[item['writer']].extend(sent_tokenize(item['text']))
print(f"Number of sentences is {len(list(itertools.chain(*(list(writers_sentences_data.values())))))}")
```

```python
writers_texts_data = {writer : [] for writer in writers}

for writer, sentences in tqdm.tqdm(writers_sentences_data.items()):
    writers_texts_data[writer].append(len(sentences) / writers_num_movies_reviewed_dict[writer])
    writers_texts_data[writer].append(np.average([len(sent) for sent in sentences]))

for writer, tokens in tqdm.tqdm(writers_texts.items()):
    writers_texts_data[writer].append(len(tokens) / writers_num_movies_reviewed_dict[writer])
    writers_texts_data[writer].append(np.average([len(word) for word in tokens]))
writers_df = pd.DataFrame(writers_texts_data, index=[
    'Average number of sentences in review', 'Average sentence length','Average number of words in review', 'Average words length'])   
```
